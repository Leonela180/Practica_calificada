```python
#edad. ,edia y promedio con barras
import numpy as np1
edad1=input("Ingrese su edad: ")
edad2=input("Ingrese su edad: ")
edad3=input("Ingrese su edad: ")
edad4=input("Ingrese su edad: ")
edad5=input("Ingrese su edad: ")
edades=[int(edad1),int(edad2),int(edad3),int(edad4),int(edad5)]

fig, ax = plt.subplots()
ax.bar([1, 2, 3, 4, 5], edades, color='green' )
```

```python
prom=sum(edades)/len(edades)
print("El promedio de edades es: \n ",prom)
media= np.mean(edades)
print("La media de edades es: \n ",media)

fig, ax = plt.subplots()
ax.bar(['Promedio', 'Media'], [prom, media])
plt.show()
```

```python
#media y desviacion estandar  con ax pie de peso en kg de 5
peso1=input("Ingrese su peso: ")
peso2=input("Ingrese su peso: ")
peso3=input("Ingrese su peso: ")
peso4=input("Ingrese su peso: ")
peso5=input("Ingrese su peso: ")
pesos=[int(peso1),int(peso2),int(peso3),int(peso4),int(peso5)]

fig, ax = plt.subplots()
ax.pie(pesos, labels=pesos)
plt.show()
```

```python

media= np.mean(pesos)
print("La media de pesos es: \n ",media)
desviacion= np.std(pesos)
print("La desviacion estandar de pesos es: \n ",desviacion)
fig, ax = plt.subplots()
ax.pie([media, desviacion], labels=['media','desviacion'])
plt.show()
```

```python
#diagrama de pastel, porcentaje de varones y mujeres
import matplotlib.pyplot as plt

mujeres=input("Ingrese el numero de mujeres: ")
varones=input("Ingrese el numero de varones: ")
total=(int(mujeres)+int(varones))
porcm=(int(mujeres)/total)*100
porcv=(int(varones)/total)*100
print("El porcentaje de mujeres es: \n ",porcm)
print("El porcentaje de varones es: \n ",porcv)
fig, ax = plt.subplots()
ax.pie([porcm, porcv], autopct='%1.1f%%')
plt.show()
```
