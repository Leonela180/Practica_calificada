```python
#Calcular la suma, resta, multiplicacion y division de dos vectores
vector1 = np.array([6, 5, 4, 2, 7])
vector2 = np.array([1, 3, 8, 9, 4])

suma = np.add(vector1,vector2)
resta = np.subtract(vector1,vector2)
multi = np.multiply(vector1,vector2)
div = np.divide(vector1,vector2)

print("Suma\n",suma)
print("Resta\n",resta)
print("Multiplicacion\n",multi)
print("Division\n",div)
```

```python
#Calcula la media, mediana y desviacion estandar de un vector
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
media = np.mean(vector1)
mediana = np.median(vector1)
desviacion = np.std(vector1)

print("Media\n",media)
print("Mediana\n",mediana)
print("Desviacion Estandar\n",desviacion)
```

```python
#Normaliza un vector dividiendo cada elemento por el valor maximo
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
maximo = np.max(vector1)
div= vector1/maximo
print("Vector Normalizado\n",div)
```

```python
#Encuentre el indice del valor maximo y min de un vector
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
maximo = np.max(vector1)
in_max = np.argmax(vector1)
minimo = np.min(vector1)
in_min = np.argmin(vector1)

print("Valor maximo\n",maximo)
print("Valor minimo\n",minimo)
print("Indice del valor maximo\n",in_max)
print("Indice del valor minimo\n",in_min)
```

```python
#Realiza una multiplicacion matricial entre dos matrices 3x3
matriz1 = np.random.randint(1,10,9).reshape(3,3)
matriz2 = np.random.randint(1,10,9).reshape(3,3)
print("Primera matriz\n",matriz1)
print("Segunda matriz\n",matriz2)
print("------------------")
multi = np.dot(matriz1,matriz2)
print("Multiplicacion\n",multi)
```

```python
import pandas as pd
```

```python
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'María', 'Jorge', 'Lucía', 'Pedro', 'Sofía', 'Raúl', 'Elena'],
    'Edad': [24, 30, 22, 28, 35, 26, 20, 40, 32, 27],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid', 'Valencia']
}

df = pd.DataFrame(data)
print("DataFrame original:")
print(df)

# 2. Filtrar a las personas con más de 25 años
filtro_mayores_25 = df[df['Edad'] > 25]
print("\nPersonas con más de 25 años:")
print(filtro_mayores_25)

# 3. Añadir nueva columna 'Ingresos'
df['Ingresos'] = [25000, 32000, 21000, 28000, 34000, 30000, 19000, 40000, 33000, 31000]
print("\nDataFrame con columna 'Ingresos':")
print(df)

# 4. Promedio de la edad
promedio_edad = df['Edad'].mean()
print("\nPromedio de edad:", promedio_edad)

# 5. Contar cuántas veces aparece cada ciudad
conteo_ciudades = df['Ciudad'].value_counts()
print("\nConteo de ciudades:")
print(conteo_ciudades)

# 6. Agrupar por ciudad y calcular promedio de edad
promedio_edad_por_ciudad = df.groupby('Ciudad')['Edad'].mean()
print("\nPromedio de edad por ciudad:")
print(promedio_edad_por_ciudad)

# 7. Ordenar por edad ascendente
df_ascendente = df.sort_values(by='Edad')
print("\nDataFrame ordenado por edad (ascendente):")
print(df_ascendente)

# 8. Ordenar por edad descendente
df_descendente = df.sort_values(by='Edad', ascending=False)
print("\nDataFrame ordenado por edad (descendente):")
print(df_descendente)
```
