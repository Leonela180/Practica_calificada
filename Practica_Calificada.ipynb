## EJERCICIO 2

```python
#Calcular la suma, resta, multiplicacion y division de dos vectores
vector1 = np.array([6, 5, 4, 2, 7])
vector2 = np.array([1, 3, 8, 9, 4])

suma = np.add(vector1,vector2)
resta = np.subtract(vector1,vector2)
multi = np.multiply(vector1,vector2)
div = np.divide(vector1,vector2)

print("Suma\n",suma)
print("Resta\n",resta)
print("Multiplicacion\n",multi)
print("Division\n",div)
```

```python
#Calcula la media, mediana y desviacion estandar de un vector
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
media = np.mean(vector1)
mediana = np.median(vector1)
desviacion = np.std(vector1)

print("Media\n",media)
print("Mediana\n",mediana)
print("Desviacion Estandar\n",desviacion)
```

```python
#Normaliza un vector dividiendo cada elemento por el valor maximo
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
maximo = np.max(vector1)
div= vector1/maximo
print("Vector Normalizado\n",div)
```

```python
#Encuentre el indice del valor maximo y min de un vector
vector1 = np.random.randint(1,10,6)
print("Vector\n",vector1)
print("------------------")
maximo = np.max(vector1)
in_max = np.argmax(vector1)
minimo = np.min(vector1)
in_min = np.argmin(vector1)

print("Valor maximo\n",maximo)
print("Valor minimo\n",minimo)
print("Indice del valor maximo\n",in_max)
print("Indice del valor minimo\n",in_min)
```

```python
#Realiza una multiplicacion matricial entre dos matrices 3x3
matriz1 = np.random.randint(1,10,9).reshape(3,3)
matriz2 = np.random.randint(1,10,9).reshape(3,3)
print("Primera matriz\n",matriz1)
print("Segunda matriz\n",matriz2)
print("------------------")
multi = np.dot(matriz1,matriz2)
print("Multiplicacion\n",multi)
```
